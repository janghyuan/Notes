## 数据准备，并导入相应的函数库 numpy 与 matplotlib.pyplot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y1 = x ** 2
y2 = 2 * x + 1
```

## 每个函数图像放在各自的窗口中

```python
# y1 函数放到第一个窗口中
plt.figure(num=1, figsize=(5, 5))
plt.plot(x, y1, color='red', linewidth=2.0, linestyle='--')

# y2 函数反倒第二个窗口中
plt.figure(num=2, figsize=(5, 5))
plt.plot(x, y2, linewidth=2.0)

# 显示两个函数图像
plt.show()
```

## 在一个窗口（figure）中显示多个函数图像

```python
plt.figure(figsize=(5, 5))

plt.plot(x, y1, color='red', linewidth=2.0)
plt.plot(x, y2, color='black', linewidth=2.0)

plt.show()
```

## 修饰坐标轴（范围、精度、文字表示、添加 label）

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y1, color='red', linewidth=2.0)
plt.plot(x, y2, color='black', linewidth=2.0)

# 只显示横坐标范围 -1 ~ 2，纵坐标范围 -2 ~ 6
plt.xlim((-1, 2))
plt.ylim((-2, 6))

# 将横坐标的范围 -1 ~ 2 进行 5 等分
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)

# 将纵坐标的标识换为文字表示
# 用 $ 包起来后，我们需要用 \ 来转义空格字符，这样做的好处就是字体变漂亮了
# \alpha \beta 用来表示拉丁字符
plt.yticks([-2, 0, 2, 4, 6], 
           [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])

# 加入 label 与 title
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.title('Fantastic Graph')

plt.show()
```

## 移动坐标轴

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y1, color='red', linewidth=2.0)
plt.plot(x, y2, color='black', linewidth=2.0)

# 只显示横坐标范围 -1 ~ 2，纵坐标范围 -2 ~ 6
plt.xlim((-1, 2))
plt.ylim((-2, 6))

# 将横坐标的范围 -1 ~ 2 进行 5 等分
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)

# 将纵坐标的标识换为文字表示
# 用 $ 包起来后，我们需要用 \ 来转义空格字符，这样做的好处就是字体变漂亮了
# \alpha \beta 用来表示拉丁字符
plt.yticks([-2, 0, 2, 4, 6], 
           [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])

# 加入 label 与 title
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.title('Fantastic Graph')

# 得到当前的坐标轴
ax = plt.gca()

# 将右边框和上边框取消掉
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 设置 bottom 为 x，设置 left 为 y
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 将 x，y 移动到 0 的位置
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.show()
```

## 添加图例

```python
plt.figure(figsize=(8, 5))

# 如果想给下面的 legend 函数中的 handles 传递参数 l1 后面必须加个逗号
l1, = plt.plot(x, y1, color='red', linewidth=2.0, label='up')
l2, = plt.plot(x, y2, color='black', linewidth=2.0, label='down')

# 只显示横坐标范围 -1 ~ 2，纵坐标范围 -2 ~ 6
plt.xlim((-1, 2))
plt.ylim((-2, 6))

# 将横坐标的范围 -1 ~ 2 进行 5 等分
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)

# 将纵坐标的标识换为文字表示
# 用 $ 包起来后，我们需要用 \ 来转义空格字符，这样做的好处就是字体变漂亮了
# \alpha \beta 用来表示拉丁字符
plt.yticks([-2, 0, 2, 4, 6], 
           [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])

# 加入 label 与 title
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.title('Fantastic Graph')

# labels 参数可以复写上面的 label 参数
plt.legend(loc='best', handles=[l1, l2,], labels=['up up', 'down down'])

plt.show()
```

## 对图上的点进行标注

```python
plt.figure(figsize=(8, 5))

plt.plot(x, y2, color='black', linewidth=1.0)

# 只显示横坐标范围 -1 ~ 2，纵坐标范围 -2 ~ 6
plt.xlim((-4, 4))
plt.ylim((-6, 6))
plt.yticks(np.arange(-6, 6, 1))
# 移动坐标轴
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# 标记要添加注释的点
x0 = 1
y0 = 2 * x0 + 1
plt.scatter(x0, y0, s=50, color='red')

# 添加一条虚线指示横坐标
# 给 plot 两个点就可以绘制一条直线，
# 主义，第一个参数 [x0, x0] 其实是两个点的分别的横坐标
plt.plot([x0,x0], [0,y0], 'k--', linewidth=2.5)
plt.plot([0, 1], [y0, y0], 'k--', linewidth=2.5)

# 添加标注
# xy参数指定给谁添加注释
# xytext指定文字的位置 textcoords指定位置相对于点的位置
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', 
             xytext=(+30, -30), textcoords='offset points', fontsize=12, 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5'))
# 前面两个参数指定文字的起始位置
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 12, 'color': 'r'})

plt.show()
```

## tick的能见度

```python
x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure()
plt.plot(x, y, linewidth=10)
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.7))

plt.show()
```

## 散点图的绘制

```python
X = np.random.normal(0, 1, 1024)
Y = np.random.normal(0, 1, 1024)
T = np.arctan2(Y, X)

plt.scatter(X, Y, c=T, s=30)
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

# 将刻度删掉
plt.xticks(())
plt.yticks(())

plt.show()
```

## 条形图的绘制

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X,+Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X,-Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x, y+0.01, '%.2f'%y, ha='center', va='bottom')
    
for x, y in zip(X, Y2):
    plt.text(x, -y-0.01, '%.2f'%-y, ha='center', va='top')

plt.ylim((-1.1, 1.1))
plt.show()
```

## 等高线图的绘制

```python
def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)

C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.1)

plt.clabel(C, inline=True, fontsize=10)

plt.show()
```

## 3D 数据的绘制

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
ax.set_zlim(-2, 2)
plt.show()
```

## 图中图的绘制

```python
fig = plt.figure()

x = [1,2,3,4,5,6,7]
y = [1,3,4,2,5,8,6]

left, bottom, width, height = 0.1, 0.1, 0.8, 0.8

ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, 'r')

left, bottom, width, height = 0.2, 0.6, 0.25, 0.25
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(y, x, 'b')

plt.axes([0.6, 0.2, 0.25, 0.25])
plt.plot(y[::-1], x, 'g')

plt.show()
```

## 次坐标轴的绘制

```python
x = np.arange(0, 10, 0.1)

y1 = 0.05 * x ** 2
y2 = -1 * y1

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(x, y1, 'g-')
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='g')

ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Y2 data', color='b')

plt.show()
```

## 动画的绘制

```python
from matplotlib import animation

fig, ax = plt.subplots()
x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x+i/10.0))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig,
                              func=animate,
                              frames=100,
                              init_func=init,
                              interval=20,
                              blit=True)

plt.show()
```
